```python
df['parental level of education'] = df['parental level of education'].astype('category')
df['parental level of education'] = df['parental level of education'].cat.codes
```

```python
df['lunch'] = df['lunch'].astype('category')
df['lunch'] = df['lunch'].cat.codes
df
```

```python
df['test preparation course'] = df['test preparation course'].astype('category')
df['test preparation course'] = df['test preparation course'].cat.codes
df
```

```python
df=df.fillna(0.0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
```

```python

df['math score'].replace('?', np.nan, inplace=True)
df['math score'] = df['math score'].astype(float)
math_score  = df['math score'].clip(0.0,100.0)
sns.histplot(math_score,bins=25,kde=True)
plt.axvline(math_score.mean(), color = 'k', linestyle='dashed', linewidth=1)
plt.show()
```

```python
df['reading score'].replace('?', np.nan, inplace=True)
df['reading score'] = df['reading score'].astype(float)
math_score  = df['reading score'].clip(0.0,100.0)
sns.histplot(math_score,bins=25,kde=True)
plt.axvline(math_score.mean(), color = 'k',linestyle='dashed', linewidth=1)
plt.show()
```
