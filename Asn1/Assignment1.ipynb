```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
del_cols = []

for col in df.columns:
    # if(type(col[4]) or type(col[5]))!= 'int':
        # count =  new_df[col].isnull().sum()
        count = df[col].isna().sum()
        percent= ((count)/len(df))*100
        print(col," ", count, " ", percent)
        if percent > 75.0:
            del_cols.append(col)
print(del_cols)
```

```python
df.drop(columns = del_cols,inplace=True)
```

```python
df.to_csv('Final.csv')
```

```python
df.drop(columns = ['Id'],inplace=True)
```

```python
df.to_csv('Final.csv')
```

```python
for col in df.columns:
    if type(df[col][4])== np.int64 or type(df[col][4]) == np.float64:
      mean_val = df[col].mean(skipna=True)
      df[col].fillna(mean_val, inplace=True)
```

```python
df.to_csv('Final.csv')
```

```python
for col in df.columns:
    if  df[col].dtype == 'O':
        df[col] = df[col].astype('category')

        # Convert categorical strings to integers using cat.codes
        df[col] = df[col].cat.codes
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = (df.drop(columns=['SalePrice'])).values
Y = df['SalePrice'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Select the top k features based on ANOVA F-statistic
k = 10  # Adjust the value of k based on your dataset
X_train_selected = SelectKBest(f_classif, k=k).fit_transform(X_train, y_train)
selected_features = SelectKBest(f_classif, k=k).fit(X_train, y_train)
X_test_selected = X_test[:, selected_features.get_support()]
```

```python
from sklearn.decomposition import PCA

# Reduce dimensionality with PCA
n_components = 10  # Adjust the number of components based on your dataset
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_pca, y_train)
```

```python
predictions = rf_classifier.predict(X_test_pca)
```

```python
acc = accuracy_score(predictions,y_test)
acc*100
```
